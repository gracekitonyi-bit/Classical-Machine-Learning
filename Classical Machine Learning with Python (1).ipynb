# Learning Styles in Machine Learning

Machine learning is essential for mimicking the natural learning process. In natural learning, knowledge is acquired through experience: the brain forms associations between inputs and outcomes, and behavior is adapted based on feedback, trial-and-error, or reinforcement.

In terms of usage of data, learning occurs in three broad categories:
* **Supervised learning** (using data with true labels)
* **Unsupervised learning** (using data without labels)
* **Semisupervised learning** (combination of previous two.)
* **Reinforcement learning** (using a predefined environment)

# Tasks in Machine Learning

By task we mean , in terms of problems to solve, there are the following:
* **Classification** (Binary or k-classification), answer in probability to yes (probabilities to $k$ classes).
* **Clustering**, answer in k clusters of data points.
* **Regression** (Linear or non linear), answer in predictions of continuous functions.
* **Feature extraction**, answer in key features in the dataset.
* **Abnomaly detection**, answer in abnormal data.
* **Inverse analysis**, answer in prediction on features from known response.

# Loss Function

In supervised machine learning, the objective is to **minimize the error** of the model on the training data. A **loss function**

$$
L: \mathcal{H} \to \mathbb{R}
$$

is a measure of how well our prediction model $f$ approximates the expected outcome by assigning a numerical loss to each hypothesis $f \in \mathcal{H}$.

The learning process can be formulated as finding the hypothesis $h^*$ that minimizes the average loss over all training examples:

$$
h^* = \arg\min_{h \in \mathcal{H}} \frac{1}{n} \sum_{i=1}^{n} l(x_i, y_i \mid h)
$$

where:  
- $x_i$ is the input feature of the $i$-th training example  
- $y_i$ is the corresponding target  
- $l(x_i, y_i \mid h)$ is the loss assigned by hypothesis $h$ on example $(x_i, y_i)$  
- $n$ is the total number of training examples


# Evaluation metrics

## Confusion matrix
This is a metric in machine learning used to determine the performance of classification model for a given set of test data.
confusion-matrix.png


**True Positive (TP)**: The actual value was positive, and the model predicted a positive value

**True Negative (TN)**:The actual value was negative, and the model predicted negative value

**False Positive**: The actual value was negative, but the model predicted positive value (the predicted value was falsely predicted, also known as type 1 error).

**False Negative**: The actual value was positive, but the model predicted negative value (the predicted value was falsely predicted, also known as type 2 error).

$$\text{Classification Accuracy} =\frac{TPs + TNs}{TPs + FPs + FNs + TNs}$$

# Mean Squared Error formula

$$\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2$$
